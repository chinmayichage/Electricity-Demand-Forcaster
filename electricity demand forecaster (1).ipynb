```python
# INITILIALIZING MODEL TRAINING 
```

```python
from xgboost import XGBRegressor
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
model_xgb = XGBRegressor(n_estimators = 1000 ,
                         early_stopping_rounds = 50,
                         learning_rate = 0.01,
                         random_state = 42,
                         objective = 'reg:squarederror')
```

```python
model_xgb.fit(X_train, Y_train, eval_set = [(X_train, Y_train),(X_test, Y_test)])
```

```python
prediction_xgb = model_xgb.predict(X_test)
```

```python
rmse_xgb = np.sqrt(mean_squared_error(Y_test, prediction_xgb))

mae_xgb = mean_squared_error(Y_test, prediction_xgb)
```

```python
print('XGBoost RMSE:', rmse_xgb)
print('XGBoost MAE:', mae_xgb)
```

```python
plt.figure(figsize = (12,6))
plt.plot(Y_test.index, Y_test, label = 'Actual Demand', color = 'Blue')
plt.plot(Y_test.index, prediction_xgb, label = 'Predicted Demand', color = 'Red', linestyle = '--')
plt.title('XGBoost Electricity Demand Prediction')
plt.xlabel("Date")
plt.ylabel("Demand (In MW)")
plt.legend()
plt.show()
```
